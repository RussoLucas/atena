```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
np.random.seed(0)
data1 = np.random.normal(0, 6, 100)
data2 = np.random.normal(0, 7, 100)
data3 = np.random.normal(0, 8, 100)
data4 = np.random.normal(0, 9, 100)

# Create violin plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.violinplot([data1, data2, data3, data4], showmeans=True, showextrema=False, showmedians=True)

# Customize plot appearance
ax.set_title('Violin Plot Example')
ax.set_xlabel('Dataset')
ax.set_ylabel('Value')

# Add labels to violin plots
ax.set_xticks([1, 2, 3, 4])
ax.set_xticklabels(['Data 1', 'Data 2', 'Data 3', 'Data 4'])

# Add grid lines
ax.yaxis.grid(True)

plt.show()
```
